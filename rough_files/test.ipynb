```python
def get_nunique():

    for i in df.columns:
        unique = df[i].nunique()
        print(f"{i} has unique values : {unique} and Nan values : {df[i].isna().sum()}")
        print(f"{i} : data type : {df[i].dtype}")

get_nunique()
    
```

```python
def object_col():

    object_column = []
    num_col = []
    for key, val in df.items():
        if pd.api.types.is_object_dtype(val):
            object_column.append(key)
        else:
            num_col.append(key)
            
    return (object_column , num_col)

object_columns, num_columns = object_col()
object_columns
```
