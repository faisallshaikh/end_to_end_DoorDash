```python
num_df = df_non_null.select_dtypes(exclude="object")
```

```python
def get_redundant(df):
    cols = set()
    for i in range(0, df.shape[1]):
        for j in range(0,i+1):
            cols.add((df.columns[i], df.columns[j]))
    return cols 
```

```python
get_redundant(num_df)
```

```python
df_non_null.columns
```

```python
num_df.columns
```

```python
def get_top_corelated_col(df,n):
    cor = df.corr(numeric_only=True).abs().unstack()
    labels_to_drop = get_redundant(df)
    a_cor = cor.drop(labels=labels_to_drop).sort_values(ascending=False)
    return a_cor[0:n]
get_top_corelated_col(num_df,20)
```
